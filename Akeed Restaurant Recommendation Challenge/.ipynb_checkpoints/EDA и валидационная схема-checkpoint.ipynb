## Описание данных

__train_customers.csv__


Информация о клиентах:


* 'akeed_customer_id': Уникальный id клиента (см. train_locations.csv и train_orders.csv)
* 'gender': пол клиента
* 'dob': год рождения
* 'status' and 'verified': статус аккаунта клиента
* 'language': язык клиента
* 'Created_at' and 'updated_at': даты создания и изменения аккаунта (created/updated)


__train_locations.csv__


Описание локаций, откуда клиенты делали заказы
* 'customer_id': Уникальный id клиента (см. __train_locations.csv__ и __train_orders.csv__)
* 'location_number': Число локаций
* 'location_type': тип локации клиента (Home, Work, Other or NA)
* 'Latitude' and 'longitude': Закодированные широта и долгота (использовались маски кодирования). Все расстояния сохранены (все локации сохранили истинное расстояние между собой). Возможны ошибки из-за неточного GPS.


__orders.csv__


Все заказы клиентов:
* 'akeed_order_id': Внутренний id (игнорируется)
* 'customer_id': ID клиента
* 'item_count': число предметов в заказе 
* 'grand_total': Общая стоимость заказа


Столбцы платежей: 
* 'payment_mode'
* 'Promo_code'
* 'vendor_discount_amount'
* 'Promo_code_discount_percentage'


Столбцы продавцов
* 'is_favorite'
* 'is_rated'
* 'vendor_rating'
* 'driver_rating'


Детали заказа:
* 'deliverydistance'
* 'preparationtime'
* 'delivery_time'
* 'order_accepted_time'
* 'driver_accepted_time'
* 'ready_for_pickup_time'
* 'picked_up_time'
* 'delivered_time'
* 'delivery_date'
* 'created_at'


* 'vendor_id': Уникальный номер продавца
* 'LOCATION_NUMBER': Номер локации доставки
* 'LOCATION_TYPE': Тип локации клиента
* 'CID X LOC_NUM X VENDOR': Customer ID, location number и Vendor number используемый в файле сабмита.


__vendors.csv__

Содержит информацию о продавцах. Важные столбцы: 
* 'id': ID продавца
* 'latitude' and 'longitude' : Широта и долгота продавца (зашифрованные)
* ‘vendor_tag_name’: Тэги-описание продавцов

### Анализ orders

```python
columns_numeric = orders.dtypes[orders.dtypes != 'O'].index

count = 0
plt.figure(figsize=(15, 10))
for idx, col in enumerate(columns_numeric):
    warnings.filterwarnings('ignore')
    
    count += 1
    plt.subplot(3, 4, count)
    plt.title(col)
    plt.hist(orders[col].values, color='g', bins=100)
    plt.axvline(x=orders[col].mean(), color='red', linestyle='--', label='mean')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Визуализация пропусков
mi.matrix(orders)
plt.title('orders 1')
plt.show()


mi.heatmap(orders)
plt.title('orders 2')
plt.show()

mi.dendrogram(orders)
plt.title('orders 3')
plt.show()
```

Рассмотрик каждый из вышенаписанных графиков `orders 1, orders 2, orders 3` подробнее:


__График orders 1__ иллюстрирует пропуски по столбцам. Полезно, когда нужно увидеть наглядно, сколько и где данных пропущено.


__График orders 2__ иллюстрирует взаимные зависимости (корреляции) между пропусками в разных столбцах. Посмотрим подробнее на этот график. 

* `delivery_date` имеет отрицательную корреляцию почти со всеми признаками. Это значит, что при наличии пропусков в этом столбце не будут пропуски в других столбцах
* `delivered_time` имеет положительную корреляцию со многими признаками. Скорее всего именно этот признак отвечает за факт доставки товара клиенту. 
* Везде, где есть значение `delivery_date`, нет значения `is_favorite`
* В оставшихся признаках интересных зависимостей не выявлено


__График orders 3__ интересен тем, что отображает, какие признаки чаще всего имеют совместные пропуски/заполнения. По нему видно, что вместе с `delivered_time` часто встречается переменная `order_accepted_time`

### Анализ train/test customers

```python
# train_customers, test_customers
columns_numeric = train_customers.dtypes[train_customers.dtypes != 'O'].index

count = 0
plt.figure(figsize=(15, 17))
for idx, col in enumerate(columns_numeric):
    warnings.filterwarnings('ignore')
    
    count += 1
    plt.subplot(3, 2, count)
    plt.title(col)
    plt.hist(train_customers[col].values, color='g', label='train', bins=100)
    plt.axvline(x=train_customers[col].mean(), color='red', linestyle='--', label='mean')
    plt.legend()
    
    
    count += 1
    plt.subplot(3, 2, count)
    plt.title(col)
    plt.hist(test_customers[col].values, color='brown', label='test', bins=100)
    plt.axvline(x=test_customers[col].mean(), color='red', linestyle='--', label='mean')
    plt.legend()

plt.tight_layout()
plt.show()
```

__!__ Видим выбросы в дате рождения как в Train, так и в Test

```python
# Визуализация пропусков
mi.matrix(train_customers)
plt.title('train_customers')
plt.show()

mi.matrix(test_customers)
plt.title('test_customers')
plt.show()
```

## Валидационная схема

```python
sample.sort_values('CID X LOC_NUM X VENDOR', inplace=True)
```

```python
def parse(val):
    return np.array([val.split(' ')[0], val.split(' ')[2], val.split(' ')[4]])

q = sample['CID X LOC_NUM X VENDOR'].apply(lambda x: parse(x))

person = [i[0] for i in q]
loc_num = [i[1] for i in q]
vendor = [i[2] for i in q]

test_dataframe = pd.DataFrame({'person': person, 'loc_num': loc_num, 'vendor': vendor})
test_dataframe.sort_values(['person', 'loc_num', 'vendor'], inplace=True)
```
